## 생선 분류 문제

### 도미 데이터 준비하기

```python
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

```python
import matplotlib.pyplot as plt

plt.scatter(bream_length, bream_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

### 빙어 데이터 준비하기

```python
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python

plt.scatter(smelt_length, smelt_weight)
plt.scatter(bream_length, bream_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

## 첫 번째 머신러닝 프로그램

```python
length = bream_length+smelt_length
weight = bream_weight+smelt_weight
```

```python
fish_data = [[l, w] for l, w in zip(length, weight)]

print(fish_data)
```

```python
fish_target = [1]*35 + [0]*14
print(fish_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier()
```

```python
kn.fit(fish_data, fish_target)
# 모델학습
```

```python
kn.score(fish_data, fish_target)
# 모델 성능 테스트 (학습시킨 데이터이기에 정확도가 100% 산출)
```

### k-최근접 이웃 알고리즘

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.scatter(30, 600, marker='^') # 길이 30 무게 600
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
kn.predict([[30, 600]]) # predict는 특정 데이터가 들어왔을 때 도미인지 빙어인지 판단하는 method
```

```python
print(kn._fit_X) # 학습된 데이터들의 y값을 출력해볼 수 있다.
```

```python
print(kn._y)
```

```python
kn49 = KNeighborsClassifier(n_neighbors=49)
```

```python
kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target) # 정확도가 떨어짐, 도미에 대한 데이터로 학습시켰지만 빙어 데이터로 성능 평가 했기에 정확도가 하향됌
```

```python
print(35/49)
```

### 확인 문제

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for n in range(5, 50):
    # 최근접 이웃 개수 설정
    kn.n_neighbors = n
    # 점수 계산
    score = kn.score(fish_data, fish_target)
    # 100% 정확도에 미치지 못하는 이웃 개수 출력
    if score < 1:
        print(n, score)
        break
```

n=18부터 떨어지는 이유는 n=17까지는 빙어 데이터로만 모델의 성능을 테스트 하였지만 n=18부터 도미 데이터가 포함되기 시작하여 정확도가 하향되었다.
